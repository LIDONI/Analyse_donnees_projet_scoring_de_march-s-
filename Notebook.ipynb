**Analyse de corrélation pour supprimer les redondances**

````python
#Créer une matrice pivot pour la corrélation
pivot_data = filtered_data.pivot(index='Country Code', columns='Indicator Code', values='2015')
# Calculer la corrélation
correlation_matrix = pivot_data.corr()
correlation_matrix
````

````python
#Heatmap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
mask = np.triu(np.ones_like(pivot_data.corr(), dtype=bool))
sns.heatmap(pivot_data.corr(), mask=mask, center=0, cmap='RdBu',
            linewidths=1, annot=True, fmt=".2f", vmin=-1, vmax=1)
plt.title('Carte des corrélations', fontsize = 15, fontweight='bold')
plt.show()
````

````python
# Identifier les indicateurs dont(corrélation > 0.7)
high_corr_pairs = set()
for i in correlation_matrix.columns:
    for j in correlation_matrix.index:
        if i != j and abs(correlation_matrix.loc[i, j]) > 0.7:
            high_corr_pairs.add(tuple(sorted((i, j))))

# Supprimer les indicateurs redondants
to_remove = set()
for pair in high_corr_pairs:
    to_remove.add(pair[1])

final_indicators = [ind for ind in top_25_indicators if ind not in to_remove] # concerver que ceux exlus de to_remove
final_indicators
````

````python
# Liste des indicateurs jugés pertinent (coefficient de corr < 0.7)
indicators_to_keep = [
    'SE.PRM.AGES',
    'SE.PRM.DURS',
    'SE.SEC.DURS.UP',
    'UIS.THDUR.0',
    'SE.SEC.DURS.LO',
    'SP.POP.TOTL',
    'SP.POP.GROW',
    'UIS.THAGE.0',
    'UIS.THAGE.3.A.GPV',
    'IT.NET.USER.P2',
    'NY.GDP.MKTP.CD',
    'NY.GDP.PCAP.CD'
]
````

````python
#Statistiques descritptives et distribution

# Fonction pour analyser un indicateur
def analyze_indicator(dataframe, indicator):

   #Statistiques descriptives
    stats = dataframe[indicator].describe()
    print(f"Statistiques descriptives pour {indicator}:\n{stats}\n")

    # Tracer la distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(dataframe[indicator], kde=True, bins=30, color='blue')
    plt.title(f"Distribution de {indicator}")
    plt.xlabel(indicator)
    plt.ylabel("Fréquence")
    plt.grid(True)
    plt.show()

# Appliquer la fonction à tous les indicateurs
for indicator in aggregated_data.columns[1:]:
    print(f" Analyse de {indicator}")
    analyze_indicator(aggregated_data, indicator)
````

````python
# Regrouper par 'Country Name' et calculer la somme de 'Normalized' pour tous les indicateurs
class_countries = (
    New_data.groupby(['Country Name'])['Normalized']
    .sum()
    .reset_index()
)

# Renommer les colonnes pour plus de clarté
class_countries.columns = ['Country Name', 'Total Normalized']

# Classer par ordre décroissant
class_countries = class_countries.sort_values(by='Total Normalized', ascending=False)
class_countries
````
